```python
import numpy as np
import pandas as pd
```

```python
data_path = '05d2b4ea-c-Dataset/'
```

```python
train_agg = pd.read_csv('05d2b4ea-c-Dataset/train_data/AggregateData_Train.csv')
train_agg.head(5)
```

```python
test_agg = pd.read_csv('05d2b4ea-c-Dataset/test_data/AggregateData_Test.csv')
test_agg.head(5)
```

### Initial Exploration

```python
print("Number of training examples: ", train_agg.shape)
print("Number of bad examples: ", train_agg[train_agg.bad_flag == 1].shape[0])
print("Percentage of bad examples: ", np.sum([train_agg.bad_flag == 1])/train_agg.shape[0]*100)
```

### Data Preparation (k-fold Cross Validation)

```python
X = train_agg.iloc[:, 2:-1].values
y = train_agg.iloc[:, -1]

X_test = test_agg.iloc[:, 2:].values

from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
i = 0
for train_idx, test_idx in skf.split(X, y):
    X_fold = X[train_idx]
    y_fold = y[train_idx]
    X_hold_out = X[test_idx]
    y_hold_out = y[test_idx]
    print(X_fold.shape, y_fold.shape, X_hold_out.shape, y_hold_out.shape, np.sum(y_hold_out))
    np.save("05d2b4ea-c-Dataset/train_split/X_fold_%d.npy"%i, X_fold)
    np.save("05d2b4ea-c-Dataset/train_split/y_fold_%d.npy"%i, y_fold)
    np.save("05d2b4ea-c-Dataset/train_split/X_hold_out_%d.npy"%i, X_hold_out)
    np.save("05d2b4ea-c-Dataset/train_split/y_hold_out_%d.npy"%i, y_hold_out)
    np.save("05d2b4ea-c-Dataset/train_split/fold_idx_%d.npy"%i, train_idx)
    np.save("05d2b4ea-c-Dataset/train_split/hold_out_idx_%d.npy"%i, test_idx)
    i += 1
    
np.save("05d2b4ea-c-Dataset/train_split/X_train.npy", X)
np.save("05d2b4ea-c-Dataset/train_split/y_train.npy", y)
np.save("05d2b4ea-c-Dataset/train_split/X_test.npy", X_test)
```
